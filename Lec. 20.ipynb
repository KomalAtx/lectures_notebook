```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/home/komal/Downloads/CardioUncleaned.csv', delimiter=';')
```

```python
data
```

```python
#data.drop(data[condition].index, axis = 0, inplace = True)

data.drop(data[data["height"]<120].index, axis = 0, inplace = True)
data.drop(data[data["weight"]<50].index, axis = 0, inplace = True)
data.drop(data[data["ap_hi"]<60].index, axis = 0, inplace = True)
data.drop(data[data["ap_hi"]>240].index, axis = 0, inplace = True)
data.drop(data[data["ap_lo"]<40].index, axis = 0, inplace = True)
data.drop(data[data["ap_lo"]>180].index, axis = 0, inplace = True)
```

```python
data.drop(['id'], axis=1,inplace=True)
```

```python
data
```

```python
#data[''] = data[''].replace( , 
data['gender'] = data['gender'].replace(1, 'female')
data['gender'] = data['gender'].replace(2, 'male')
data['cholesterol'] = data['cholesterol'].replace(1, 'chl_avg')
data['cholesterol'] = data['cholesterol'].replace(2, 'chl_aboveavg')
data['cholesterol'] = data['cholesterol'].replace(3, 'chl_high')
data['gluc'] = data['gluc'].replace(1, 'glc_avg')
data['gluc'] = data['gluc'].replace(2, 'glc_aboveavg')
data['gluc'] = data['gluc'].replace(3, 'glc_high')
```

```python
data
```

```python
data.to_csv('cardiocleaned.csv', index = None)     #saving data in a few file in csv format
```

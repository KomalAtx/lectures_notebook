```python
from nltk.corpus import stopwords
import pandas as pd
```

```python
Data = pd.read_csv('AmazonMobileDataCleaned.csv')
```

```python
Data
```

```python
stop = set(stopwords.words('english'))   #set have unique elements 
```

```python
stop
```

```python
stop.remove('no')
```

```python
stop.remove('not')
```

```python
cleaned2 = []   #list of cleaned reviews
```

```python
for feedback in (Data['cleanedreview'].values):
    x = feedback.split()  #gives a list of words contained in feedback 
    #eg. x = ['this', 'mobile', 'is', 'good']
    cleaned_feedback = ' '.join(e for e in x if e not in stop)    #joining the words to a string except stopwords
    #eg. cleaned_feedback = 'mobile good'
    cleaned2.append(cleaned_feedback.lower().strip())    #appending the string
```

```python
Data['cleanedreview_2'] = cleaned2
```

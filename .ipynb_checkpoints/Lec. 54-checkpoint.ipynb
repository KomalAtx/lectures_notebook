```python
import numpy as np
from numpy import log, dot, exp, shape, sum
from scipy.stats import logistic
```

```python
def fit(x,y):
    lr = 0.01
    n_iters = 1000
    num_samples, num_features = x.shape             #X = [m x (n+1)],   Y = [m x 1]
    # print(type(x))
    # print(x.shape)
    weights = np.random.rand(shape(x)[1]+1,1)          #W = [(n+1) x 1],   p = [m x 1],  b = [m x 1]
    b = np.ones((shape(x)[0],1))                  #b = bias
    X = np.c_[b,x]   
    # print("X shape is :", X.shape, "weights shape is : " , weights.shape, X)
                                 # p = [m x 1]
    z= np.dot(X,weights)                               # z = [m x 1] 
    p = 1/(1+ np.exp(-z))
    # print("weights 1st element is : ", weights[0] )
    print(y.shape, p.shape,z)
    epsilon = 1e-10
    for i in range(n_iters):
        
        #Loss
        A =  np.dot(y.T,np.log(p + epsilon))
        B = np.dot((1-y).T,np.log(1-p + epsilon))
        Loss = -1 * (A + B)
        Loss = Loss/num_samples
        
        
        dw = np.dot(X.T,(p-y)) / num_samples                         #dw = [(n+1) x 1]
        weights = weights - lr * dw

        z= np.dot(X,weights)                               # z = [m x 1] 
        p = 1/(1+ np.exp(-z)) 
        # print('---->',p,A,B)
        print("weights 1st element is : ", weights[1]  , "Loss at iteration " , i, " : ", Loss)

    print(weights)
```

```python
fit(input_train, outputtrain)
```
